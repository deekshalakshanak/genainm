```python
import gradio as gr
from transformers import pipeline
import math

# Load a Hugging Face text classification model
classifier = pipeline("text-classification", model="distilbert-base-uncased-finetuned-sst-2-english")

def calculate_perplexity(text):
    """Calculates perplexity based on model confidence."""
    result = classifier(text)
    score = result[0]['score']
    # Perplexity formula: lower is better
    perplexity = math.pow(2, -math.log2(score))
    return round(perplexity, 2)

# Define the prediction function with enhanced sentiment analysis, GIF display, and perplexity

def classify_text(text):
    result = classifier(text)
    label = result[0]['label']
    score = result[0]['score']
    perplexity = calculate_perplexity(text)

    # Define colors, emojis, and GIF URLs for different sentiments
    if label == "POSITIVE":
        color = "green"
        emoji = " 👍"
        gif_url = "https://media.giphy.com/media/26AHONQ79FdWZhAI0/giphy.gif"
    elif label == "NEGATIVE":
        color = "red"
        emoji = " 👎"
        gif_url = "https://media.giphy.com/media/3o6ZsYm5cRaC0FnuXe/giphy.gif"
    else:
        color = "yellow"
        emoji = " 😐"
        gif_url = "https://media.giphy.com/media/l3V0megwbBeETMgGA/giphy.gif"

    # Create a summary
    summary = f"The text is classified as <span style='color:{color}; font-weight:bold;'>{label}</span> with a confidence score of {score:.2f}. Perplexity: {perplexity}"

    # Return formatted result, score, GIF URL, and perplexity
    result_display = f"<span style='color:{color}; font-weight:bold;'>Label: {label}{emoji}, Confidence: {score:.2f}</span><br><br>{summary}"
    return result_display, score, gif_url, perplexity

# Set up the Gradio interface with enhanced styling, a progress bar, and GIF display
interface = gr.Interface(
    fn=classify_text,
    inputs=gr.Textbox(placeholder="Enter your text here...", label="Your Text", lines=3),
    outputs=[
        gr.HTML(label="Classification Result"),
        gr.Slider(minimum=0, maximum=1, step=0.01, label="Confidence Score", interactive=False),
        gr.Image(label="Sentiment GIF"),
        gr.Number(label="Perplexity (Lower is Better)", precision=2)
    ],
    title="✨ Sentiment Classifier App ✨",
    description="Enter any text below to analyze its sentiment using a pre-trained Hugging Face model. Now with perplexity analysis!",
    theme="compact",
    css=".container {background-color: #000000; font-family: 'Arial', sans-serif; padding: 20px; border-radius: 15px; color: #ffffff;} .interface {box-shadow: 0px 4px 15px rgba(0, 0, 0, 0.2); border-radius: 12px;} textarea {border-radius: 8px; padding: 10px; color: #ffffff; background-color: #333333;} .output {background-color: #444444; border-radius: 8px; padding: 10px; color: #ffffff; text-align:center;}"
)

# Launch the app
interface.launch()
```

```python
import gradio as gr
from transformers import pipeline
import math
from textblob import TextBlob

# Load Hugging Face models for sentiment analysis, summarization, and emotion detection
classifier = pipeline("text-classification", model="distilbert-base-uncased-finetuned-sst-2-english")
summarizer = pipeline("summarization", model="t5-small")
emotion_detector = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base")


def calculate_perplexity(text):
    """Calculates perplexity based on model confidence."""
    result = classifier(text)
    score = result[0]['score']
    perplexity = math.pow(2, -math.log2(score))
    return round(perplexity, 2)


def check_grammar_and_spelling(text):
    """Performs grammar and spelling correction."""
    blob = TextBlob(text)
    corrected_text = str(blob.correct())
    return corrected_text


def determine_sentiment(label, text):
    """Handles mixed sentiment cases."""
    if "bad" in text.lower() or "terrible" in text.lower():
        return "NEGATIVE", "red", "👎", "https://media.giphy.com/media/3o6ZsYm5WJnTQp5KuY/giphy.gif"
    elif label == "POSITIVE":
        return "POSITIVE", "green", "👍", "https://media.giphy.com/media/5GoVLqeAOo6PK/giphy.gif"
    else:
        return "NEUTRAL", "blue", "😐", "https://media.giphy.com/media/26AHONQ79FdWZhAI0/giphy.gif"


def detect_emotion(text):
    """Detects emotions in the given text."""
    emotions = emotion_detector(text)
    top_emotion = max(emotions, key=lambda x: x['score'])
    return top_emotion['label'], round(top_emotion['score'], 2)


def classify_text(text):
    corrected_text = check_grammar_and_spelling(text)

    result = classifier(corrected_text)
    label = result[0]['label']
    score = result[0]['score']
    perplexity = calculate_perplexity(corrected_text)

    label, color, emoji, gif_url = determine_sentiment(label, corrected_text)

    if len(corrected_text.split()) < 20:
        summary = "Text is too short for summarization."
    else:
        summary_result = summarizer(corrected_text, max_length=50, min_length=10, do_sample=False)
        summary = summary_result[0]['summary_text']

    emotion, emotion_score = detect_emotion(corrected_text)

    result_display = f"<span style='color:{color}; font-weight:bold;'>Label: {label}{emoji}, Confidence: {score:.2f}</span><br><br>The text is classified as <span style='color:{color}; font-weight:bold;'>{label}</span> with a confidence score of {score:.2f}. Perplexity: {perplexity}<br><br><span style='color:orange; font-weight:bold;'>Emotion Detected: {emotion} (Confidence: {emotion_score:.2f})</span>"
    return result_display, summary, score, gif_url, perplexity, corrected_text


interface = gr.Interface(
    fn=classify_text,
    inputs=gr.Textbox(placeholder="Enter your text here...", label="Your Text", lines=3),
    outputs=[
        gr.HTML(label="Classification Result"),
        gr.Textbox(label="Summary", interactive=False),
        gr.Slider(minimum=0, maximum=1, step=0.01, label="Confidence Score", interactive=False),
        gr.Image(label="Sentiment GIF"),
        gr.Number(label="Perplexity (Lower is Better)", precision=2),
        gr.Textbox(label="Corrected Text", interactive=False)
    ],
    title="✨ Sentiment, Summarization, Emotion & Grammar Checker Web App ✨",
    description="Enter any text below to analyze its sentiment, get a summary, detect emotions, correct grammar, and see its perplexity!",
    theme="compact",
    css=".container {background-color: #000000; font-family: 'Arial', sans-serif; padding: 20px; border-radius: 15px; color: #ffffff;} .interface {box-shadow: 0px 4px 15px rgba(0, 0, 0, 0.2); border-radius: 12px;} textarea {border-radius: 8px; padding: 10px; color: #00ffff; background-color: #333333; transition: all 0.3s ease-in-out;} textarea:focus {transform: scale(1.05); box-shadow: 0 0 10px #00ffff;} .output {background-color: #444444; border-radius: 8px; padding: 10px; color: #ffffff; text-align:center;} .gr-number input {background-color: #222222; color: #00ff00;} .gr-number label {color: #ffffff; font-weight: bold;}"
)

interface.launch()
```
